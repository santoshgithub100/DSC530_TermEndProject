```python
"""
Term end project on covid 19 public health data for seattle & king county. 
The dataset used in this termend project is just a sample using the key metrices to analyze the COVID 19 Virus spread. The Dataset has 8000 plus rows and automated via using 
an automation scripts.

"""

from __future__ import print_function

import csv
import logging
import sys
import numpy as np
import pandas as pd

import thinkplot
import thinkstats2
import datetime
import regression
import statsmodels.formula.api as smf
import statsmodels.api as sm


def Mode(hist):
    """Returns the value with the highest frequency.
    hist: Hist object
    returns: value from Hist
    """
    p, x = max([(p, x) for x, p in hist.Items()])
    return x

# method to read the project data set.
def ReadData(filename='COVID19_KingCounty_Data_ByWeek.csv'):
    """Reads filename and returns project data in thousands

    filename: string

    returns: pandas Series of projectdata in thousands
    """
    projectData = pd.read_csv(filename, header=None, skiprows=2,
                         encoding='iso-8859-1')
    return projectData


def Execute_Populations_Variable(data):
    # to read the population Data
    populations = data[1]
    populations.replace(0, np.nan, inplace=True)
    
    populations_Hist = thinkstats2.Hist(populations);
    print('population histogram', populations_Hist);
    thinkplot.Hist(populations_Hist)
    
    # calculate Mode for population Field    
    mode = Mode(populations_Hist)
    print('The mode value for population field is : ', mode)

    # calculate Mean for population Field
    log_pops = np.log10(populations)
    mean = log_pops.mean()
    print('The mean value for population field is : ', mean)

    # calculate Variance for population Field
    var = log_pops.var()
    print('The Variance for population field is : ', var)
    
    # calculate Standard Deviation for population Field
    std = log_pops.std()
    print('The Standard Deviation for population field is : ', std)

    
def Execute_casecounts_Variable(data):
    # to read the case counts Data
    casecounts = data[4]
    casecounts.replace(0, np.nan, inplace=True)
    
    casecounts_Hist = thinkstats2.Hist(casecounts);
    print('case_counts_Hist', casecounts_Hist);
    
    thinkplot.Hist(casecounts_Hist)
    
    # calculate Mode for case counts Field    
    mode = Mode(casecounts_Hist)
    print('The mode value for case counts field is : ', mode)

    # calculate Mean for case counts Field 
    log_casecounts = np.log10(casecounts)
    mean = log_casecounts.mean()
    print('The mean value for case counts field is : ', mean)

     # calculate Variance for case counts Field
    var = log_casecounts.var()
    print('The Variance for case counts field is : ', var)
    
    # calculate Standard Deviation for case counts Field
    std = log_casecounts.std()
    print('The Standard Deviation for case counts field is : ', std)


def Execute_hospitalcounts_Variable(data):
    # to read the hospital counts Data
    hospitalcounts = data[5]
    hospitalcounts.replace(0, np.nan, inplace=True)
    
    hospitalcounts_hist = thinkstats2.Hist(hospitalcounts);
    print('hospital_counts_Hist', hospitalcounts_hist);
    
    thinkplot.Hist(hospitalcounts_hist)
    
    # calculate Mode for hospital Field 
    mode = Mode(hospitalcounts_hist)
    print('The mode value for hospital counts field is : ', mode)

    # calculate Mean for hospital Field 
    log_hospitalcounts = np.log10(hospitalcounts)
    mean = log_hospitalcounts.mean()
    print('The mean value for hospital counts field is : ', mean)

    # calculate Variance for hospital counts Field
    var = log_hospitalcounts.var()
    print('The Variance for case counts field is : ', var)
    
    # calculate Standard Deviation for hospital counts Field
    std = log_hospitalcounts.std()
    print('The Standard Deviation for case counts field is : ', std)
    
def Execute_deathcounts_Variable(data):
    # to read the death counts Data
    deathcounts = data[6]
    deathcounts.replace(0, np.nan, inplace=True)
    
    deathcounts_hist = thinkstats2.Hist(deathcounts);
    print('death_counts_Hist', deathcounts_hist);
    
    thinkplot.Hist(deathcounts_hist)
    
    # calculate Mode for death Field 
    mode = Mode(deathcounts_hist)
    print('The mode value for death counts field is : ', mode)

    # calculate Mean for death Field 
    log_deathcount = np.log10(deathcounts)
    mean = log_deathcount.mean()
    print('The mean value for death counts field is : ', mean)

    # calculate Variance for death counts Field
    var = log_deathcount.var()
    print('The Variance for death counts field is : ', var)
    
    # calculate Standard Deviation for death counts Field
    std = log_deathcount.std()
    print('The Standard Deviation for death counts field is : ', std)

def Execute_pcrtestcounts_Variable(data):
    # to read the pcrtest counts Data
    pcrtestcounts = data[8]
    pcrtestcounts.replace(0, np.nan, inplace=True)
    
    pcrtestcounts_hist = thinkstats2.Hist(pcrtestcounts);
    print('death_counts_Hist', pcrtestcounts_hist);
    
    thinkplot.Hist(pcrtestcounts_hist)
    
    # calculate Mode for pcrtest Field 
    mode = Mode(pcrtestcounts_hist)
    print('The mode value for pcrtest counts field is : ', mode)

    # calculate Mean for pcrtest Field 
    #log_pcrtestcounts = np.log10(pcrtestcounts)
    mean = pcrtestcounts.mean()
    print('The mean value for pcrtest counts field is : ', mean)

    # calculate Variance for pcr test counts Field
    var = pcrtestcounts.var()
    print('The Variance for pcr test counts field is : ', var)
    
    # calculate Standard Deviation for pcr test counts Field
    std = pcrtestcounts.std()
    print('The Standard Deviation for pct test counts field is : ', std)


def Execute_HospitalizationsPmfVar(data):

     df = pd.DataFrame(data)
     df[2] = pd.to_datetime(df[2])

     #print (df[2])

     firstdaterange = df[(df[2] > "2020-01-05") & (df[2] < "2020-02-02")]
     #print(firstdaterange)
     print("date > 2020-01-05 and date < 2020-02-02 : No Of Rows :", len(firstdaterange))

     hospitalization_firstRange_dropna = firstdaterange[5].dropna()

     seconddaterange = df[(df[2] > "2020-10-26") & (df[2] < "2021-04-04")]
     print("date > 2020-10-26 and date < 2021-04-04 : No Of Rows :", len(seconddaterange))

     hospitalization_secondRange_dropna = seconddaterange[5].dropna()

     hospitalization_first_dateRange_pmf = thinkstats2.Pmf(hospitalization_firstRange_dropna, label='Hospitalization [Week_Start > 2020-01-05 and Week_Start < 2020-02-02')
     hospitalization_second_dateRange_pmf = thinkstats2.Pmf(hospitalization_secondRange_dropna, label='Hospitalization [Week_Start > 2020-10-26 and Week_Start < 2021-04-04]')

     width = 0.4 / 2

     # plot PMFs of birth weights for first babies and others
     thinkplot.PrePlot(2)
     thinkplot.Hist(hospitalization_first_dateRange_pmf, align='right', width=width)
     thinkplot.Hist(hospitalization_second_dateRange_pmf, align='left', width=width)
     thinkplot.Config(xlabel='Hospitalization Count (numbers)', ylabel='PMF')



def Execute_DeathsPmfVar(data):
     df = pd.DataFrame(data)
     df[2] = pd.to_datetime(df[2])

     #print (df[2])

     firstdaterange = df[(df[2] > "2019-01-02") & (df[2] < "2021-01-02")]
     #print(firstdaterange)
     print("date > 2019-01-02 and date < 2021-01-02 : No Of Rows :", len(firstdaterange))

     deaths_firstRange_dropna = firstdaterange[6].dropna()

     seconddaterange = df[(df[2] > "2021-10-10") & (df[2] < "2022-05-05")]
     print("date > 2021-10-10 and date < 2022-05-05 : No Of Rows :", len(seconddaterange))

     deaths_secondRange_dropna = seconddaterange[6].dropna()

     deaths_first_dateRange_pmf = thinkstats2.Pmf(deaths_firstRange_dropna, label='Deaths [Week_Start > 2019-01-02 and Week_Start < 2021-01-02')
     deaths_second_dateRange_pmf = thinkstats2.Pmf(deaths_secondRange_dropna, label='Deaths [Week_Start > 2021-10-10 and Week_Start < 2022-05-05]')

     width = 0.4 / 2

     # plot PMFs of birth weights for first babies and others
     thinkplot.PrePlot(2)
     thinkplot.Hist(deaths_first_dateRange_pmf, align='right', width=width)
     thinkplot.Hist(deaths_second_dateRange_pmf, align='left', width=width)
     thinkplot.Config(xlabel='Deaths Count (numbers)', ylabel='PMF')


def Execute_PCRTestCDF(data):

    cdf = thinkstats2.Cdf(data[8], label='pcr test')
    thinkplot.Cdf(cdf)
    thinkplot.Config(xlabel='pcr test (weeks)', ylabel='CDF', loc='upper left')

    # Cdf provides `Prob`, which evaluates the CDF; that is, it computes the fraction of values less than or equal to the given value.  For example, 14% of pcr test are less than or equal to 41.
    print("cdf.Prob percentage :", cdf.Prob(200))

    # Value evaluates the inverse CDF; given a fraction, it computes the corresponding value.  For example, the median is the value that corresponds to 0.6.
    print("cdf values :", cdf.Value(0.6))


def Execute_AnalysticalDistribution_Hospitalization(data):
    #bins = np.arange(10, 44, 3)
    #indices = np.digitize(data[6], bins)
    #groups = data.groupby(indices)

    hospitalization = [data[6].mean() for i, group in data][1:-1]
    cdfs = [thinkstats2.Cdf(data[6]) for i, group in data][1:-1]

    thinkplot.PrePlot(3)
    for percent in [75, 50, 25]:
        pcr_test = [cdf.Percentile(percent) for cdf in cdfs]
        label = '%dth' % percent
        thinkplot.Plot(hospitalization, pcr_test, label=label)

    thinkplot.config(xlabel="hospitalization's (Numbers)",
                   ylabel='PCR Test (Numbers)')
    

def ScatterPlot(data, alpha=1.0):
    """Make a scatter plot and save it. """

    hospitalization = data[5]
    deaths = data[6]

    thinkplot.Scatter(hospitalization, deaths, alpha=alpha)
    thinkplot.Config(xlabel='hospitalization',
                     ylabel='deaths',
                     legend=False)


def AnalyticalDitribution_PCRTest_ParetoModel(data):
    pcrtest = data[8]
    print("count of total PCR test taken : ", len(pcrtest))

    log_pcrtest = np.log10(pcrtest)
    cdf = thinkstats2.Cdf(pcrtest, label="data")
    cdf_log = thinkstats2.Cdf(log_pcrtest, label="data")

    # pareto plot
    xs, ys = thinkstats2.RenderParetoCdf(xmin=5000, alpha=1.4, low=0, high=1e7)
    thinkplot.Plot(np.log10(xs), 1 - ys, label="model", color="0.8")

    thinkplot.Cdf(cdf_log, complement=True)
    thinkplot.Config( xlabel="log10 pcr test", ylabel="CCDF", yscale="log", loc="lower left")

    thinkstats2.NormalProbabilityPlot(log_pcrtest, label="data")
    thinkplot.Config(xlabel="Random variate", ylabel="log10 PCR test", xlim=[-5, 5])


class DiffMeansPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        group1, group2 = data
        test_stat = abs(group1.mean() - group2.mean())
        return test_stat

    def MakeModel(self):
        group1, group2 = self.data
        self.n, self.m = len(group1), len(group2)
        self.pool = np.hstack((group1, group2))

    def RunModel(self):
        np.random.shuffle(self.pool)
        data = self.pool[:self.n], self.pool[self.n:]
        return data
    
class DiffMeansResample(DiffMeansPermute):
    """Tests a difference in means using resampling."""
    
    def RunModel(self):
        """Run the model of the null hypothesis.

        returns: simulated data
        """
        group1 = np.random.choice(self.pool, self.n, replace=True)
        group2 = np.random.choice(self.pool, self.m, replace=True)
        return group1, group2
    
def RunHypothesisTest(data):
    """Tests differences in means by resampling.

    deaths_first_range : DataFrame
    deaths_second_range : DataFrame
    """

    df = pd.DataFrame(data)
    df[2] = pd.to_datetime(df[2])

    # Filtering death data in the date range between 2019-01-02 and 2021-01-02
    firstdaterange = df[(df[2] > "2019-01-02") & (df[2] < "2021-01-02")]
    deaths_first_range = firstdaterange[6]

    # Filtering death data in the date range between 2021-10-10 and 2022-05-05
    seconddaterange = df[(df[2] > "2021-10-10") & (df[2] < "2022-05-05")]
    deaths_second_range = seconddaterange[6]

    data = deaths_first_range, deaths_second_range
    ht = DiffMeansResample(data)
    p_value = ht.PValue(iters=10000)
    print('\ndiff means resample COVID 19 Deaths')
    print('p-value =', p_value)
    print('actual =', ht.actual)
    print('ts max =', ht.MaxTestStat())

    data = (deaths_first_range.dropna().values, 
            deaths_second_range.dropna().values)
        
    ht = DiffMeansPermute(data)
    p_value = ht.PValue(iters=10000)
    print('\ndiff means resample COVID 19 Deaths')
    print('p-value =', p_value)
    print('actual =', ht.actual)
    print('ts max =', ht.MaxTestStat())

def RunMultipleRegression(data):
    
    df = pd.DataFrame(data) 

    # X axis on case_counts and Hospital_counts
    x = df[[4,5]]

    # Y axis on death_counts
    y = df[6]
    x = sm.add_constant(x)

    model = sm.OLS(y, x).fit()
    predictions = model.predict(x) 

    print_model = model.summary()
    print(print_model)

def main():
    
   data = ReadData();

   # Include the other descriptive characteristics about the variables: Mean, Mode, Spread, and Tails (Chapter 2).
   # to find mean, variance, mode and standard Deviations for 5 numeric Variable from the project data set.
   Execute_Populations_Variable(data)
   Execute_casecounts_Variable(data)
   Execute_hospitalcounts_Variable(data)
   Execute_deathcounts_Variable(data)
   Execute_pcrtestcounts_Variable(data)

   # Scenario 1: compute the distribution of hospitalization counts data due to COVID-19 pendamic for [Week_Start > 2020-01-05 and Week_Start < 2020-02-02] and [Week_Start > 2020-10-26 and Week_Start < 2021-04-04] and then plot the PMFs on the same scale
   Execute_HospitalizationsPmfVar(data)

   # Scenario 2: compute the distribution of death counts data due to COVID-19 pendamic for [Week_End > 2019-01-02 and Week_End < 2021-01-02] and [Week_End > 2021-10-10 and Week_End < 2022-05-05] and then plot the PMFs on the same scale
   Execute_DeathsPmfVar(data)

   # Create 1 CDF with one of your variables, using page 41-44 as your guide, what does this tell you about your variable and how does it address the question you are trying to answer (Chapter 4)
   Execute_PCRTestCDF(data) 

   Execute_AnalysticalDistribution_Hospitalization(data)

   # Create two scatter plots comparing two variables and provide your analysis on correlation and causation. Remember, covariance, Pearson’s correlation, and Non-Linear Relationships 
   # should also be considered during your analysis (Chapter 7).
   ScatterPlot(data, alpha=0.1)

   # Plot 1 analytical distribution and provide your analysis on how it applies to the dataset you have chosen (Chapter 5).
   AnalyticalDitribution_PCRTest_ParetoModel(data)

   # Conduct a test on your hypothesis using one of the methods covered in Chapter 9.
   RunHypothesisTest(data)

   # For this project, conduct a regression analysis on either one dependent and one explanatory variable, or multiple explanatory variables (Chapter 10 & 11).
   RunMultipleRegression(data)
   

if __name__ == "__main__":
    main()
    
```
